## Question 1
Import and load the Fashion MNIST dataset. What are the dimensions of train_images, train_labels, test_images, and test_labels?

```python
from keras.datasets import fashion_mnist

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

print('train_images shape:', train_images.shape)
print('train_labels shape:', train_labels.shape)
print('test_images shape:', test_images.shape)
print('test_labels shape:', test_labels.shape)
```

## Question 10
Import models and layers from keras.

```python
from keras import models
from keras import layers
```

## Question 11
Define a sequential model called myNetwork.

```python
myNetwork = models.Sequential()
```

## Question 12
Reshape (flatten) images from 28x28 to 784 and normalize by dividing by 255.

```python
# Reshape from 28x28 to 784 (flatten)
train_images_flat = train_images.reshape((60000, 28 * 28))
test_images_flat = test_images.reshape((10000, 28 * 28))

# Normalize by dividing by 255
train_images_flat = train_images_flat.astype('float32') / 255
test_images_flat = test_images_flat.astype('float32') / 255

print('Reshaped train_images shape:', train_images_flat.shape)
print('Reshaped test_images shape:', test_images_flat.shape)
```

## Question 13
Add a hidden layer with 512 neurons using relu activation.

```python
myNetwork.add(layers.Dense(512, activation='relu', input_shape=(784,)))
```

## Question 14
Add another hidden layer with 128 neurons using relu activation.

```python
myNetwork.add(layers.Dense(128, activation='relu'))
```

## Question 15
Add the output layer with 10 neurons using softmax activation.

**Why softmax?** Softmax converts the raw output scores (logits) into probabilities that sum to 1.0. Each neuron outputs the probability of the input belonging to that class. The class with the highest probability is the prediction. Under the hood, softmax computes e^(zi) / sum(e^(zj)) for each class i, which ensures all outputs are positive and normalized.

```python
myNetwork.add(layers.Dense(10, activation='softmax'))
myNetwork.summary()
```

## Question 16 - Option A
Compile with adam optimizer and sparse_categorical_crossentropy loss.

```python
# Option A: adam + sparse_categorical_crossentropy
myNetwork.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

# Train with 10 epochs
history_a = myNetwork.fit(train_images_flat, train_labels, epochs=10)

print('\nOption A - Final training accuracy:', history_a.history['accuracy'][-1])
```

## Question 16 - Option B
Compile with rmsprop optimizer and categorical_crossentropy loss.

Note: categorical_crossentropy requires one-hot encoded labels, so we convert them using to_categorical.

```python
from keras.utils import to_categorical

# One-hot encode labels for categorical_crossentropy
train_labels_onehot = to_categorical(train_labels)
test_labels_onehot = to_categorical(test_labels)

# Build a new model for Option B
myNetwork_b = models.Sequential()
myNetwork_b.add(layers.Dense(512, activation='relu', input_shape=(784,)))
myNetwork_b.add(layers.Dense(128, activation='relu'))
myNetwork_b.add(layers.Dense(10, activation='softmax'))

# Option B: rmsprop + categorical_crossentropy
myNetwork_b.compile(optimizer='rmsprop',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])

# Train with 10 epochs
history_b = myNetwork_b.fit(train_images_flat, train_labels_onehot, epochs=10)

print('\nOption B - Final training accuracy:', history_b.history['accuracy'][-1])
```

## Question 17
Compare Option A and Option B training results.

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_a.history['accuracy'], label='Option A (adam)')
plt.plot(history_b.history['accuracy'], label='Option B (rmsprop)')
plt.title('Training Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_a.history['loss'], label='Option A (adam)')
plt.plot(history_b.history['loss'], label='Option B (rmsprop)')
plt.title('Training Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

## Question 18
Evaluate both models on the test set to check for overfitting.

```python
# Evaluate Option A
test_loss_a, test_acc_a = myNetwork.evaluate(test_images_flat, test_labels)
print(f'Option A - Test accuracy: {test_acc_a:.4f}, Test loss: {test_loss_a:.4f}')
print(f'Option A - Train accuracy: {history_a.history["accuracy"][-1]:.4f}')
print(f'Option A - Difference (train - test): {history_a.history["accuracy"][-1] - test_acc_a:.4f}')
print()

# Evaluate Option B
test_loss_b, test_acc_b = myNetwork_b.evaluate(test_images_flat, test_labels_onehot)
print(f'Option B - Test accuracy: {test_acc_b:.4f}, Test loss: {test_loss_b:.4f}')
print(f'Option B - Train accuracy: {history_b.history["accuracy"][-1]:.4f}')
print(f'Option B - Difference (train - test): {history_b.history["accuracy"][-1] - test_acc_b:.4f}')
print()
print('If training accuracy is significantly higher than test accuracy, we have overfitting.')
print('Fashion MNIST typically achieves ~88-89% test accuracy, lower than MNIST (~97-98%)')
print('because clothing items are more visually complex and harder to distinguish than digits.')
```
